```python
import matplotlib.pyplot as plt   
import seaborn as sns
import folium as fo
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('variable_geo.csv')
```

```python
coordenadas = pd.read_excel('shops_en2.xls')
```

```python
coordenadas = coordenadas.coor.str.split(expand=True)
coordenadas.columns=['lat','long']
coordenadas.lat = coordenadas.lat.str.replace(',', '')
```

## Situamos las tiendas del dataset en el mapa Rusia

```python
# Russia
latitude = 61.52401
longitude = 105.318756
russia_map = fo.Map(location=[latitude, longitude], zoom_start=2)
shops = fo.map.FeatureGroup()
for lat, lng, in zip(coordenadas.lat, coordenadas.long):
    shops.add_child(
        fo.CircleMarker(
            [lat, lng],
            radius=6, # difine el tamaño de los circulos
            color='blue',
            fill=True,
            fill_color='yellow',
            fill_opacity=0.6
        )
    )

russia_map.add_child(shops)
for lat, lng, label in zip(coordenadas.lat, coordenadas.long, df.shop_name):
    fo.Marker([lat,lng], popup=label).add_to(russia_map)
russia_map
    
```

Al situar las tiendas en el mapa podemos observar que se encuentran muy repartidas por el territorio Ruso, por los que nos hace sospehcar que el compartamiento del cliente, va ha ser muy diferente en las distintas zonas, entendiendo que por ejemplo, que variables tipo climatológicas o económicas, tendrán una gran importancia en el modelo.
